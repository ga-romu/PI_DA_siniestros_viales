Calcular tasa de valores 'SD' en columnas, ya que el conteo de valores nulos no entregan suficiente información sobre distribución de datos en columnas. 

```python
import re

def procesar_otra_direccion(direccion):
  """
  Procesa la información de la columna "otra_direccion" y 
  rellena las columnas "cruce", "calle" y "tipo_calle".

  Args:
    direccion: String con la información de la dirección.

  Returns:
    Diccionario con los valores para las columnas "cruce", "calle" y "tipo_calle".
  """
  cruce = False
  calle = ""
  tipo_calle = "CALLE"

  # Detectar cruce
  if " y " in direccion:
    # Se busca la primera ocurrencia de "y" que no esté entre paréntesis.
    match = re.search(r"y(?!\([^)]*\))", direccion)
    if match:
      cruce = True
      calle_sin_numero = direccion[:match.start()].strip()
    else:
      calle_sin_numero = direccion.strip()
  else:
    calle_sin_numero = direccion.strip()

  # Obtener nombre de la calle
  # Se utiliza una expresión regular más robusta para identificar el final del nombre de la calle.
  match = re.search(r"(?<!\d)\b\d+\b(,\s*,|,\s*Barrio:|,\s*Altura:|,\s*Piso:|,\s*Dto:|,\s*Localidad:|,\s*Provincia:|,\s*Partido:|,\s*Código Postal:|\s*$)", calle_sin_numero)
  if match:
    calle = calle_sin_numero[:match.start()].strip()
  else:
    calle = calle_sin_numero

  # Detectar tipo de calle
  if "AU" in calle:
    tipo_calle = "AUTOPISTA"
  elif "Av." in direccion or "av." in direccion or "avenida" in direccion:
    if "Paz, Gral. Av." in direccion:
      tipo_calle = "GRAL PAZ"
    else:
      tipo_calle = "AVENIDA"

  return {"cruce": cruce, "calle": calle, "tipo_calle": tipo_calle}
```

```python
lesiones_hechos["cruce"] = lesiones_hechos["otra_direccion"].apply(lambda x: procesar_otra_direccion(x)["cruce"])
lesiones_hechos["calle"] = lesiones_hechos["otra_direccion"].apply(lambda x: procesar_otra_direccion(x)["calle"])
lesiones_hechos["tipo_calle"] = lesiones_hechos["otra_direccion"].apply(lambda x: procesar_otra_direccion(x)["tipo_calle"])
```

Procedemos a normalizar columna sexo con valores: FEMENINO, MASCULINO, SD

```python
def clean_data(lesiones_victimas):
    # Replace all instances of "Varon" with "MASCULINO" in column: 'Sexo'
    lesiones_victimas.loc[lesiones_victimas['Sexo'].str.lower() == "Varon".lower(), 'Sexo'] = "MASCULINO"
    # Replace all instances of "varon" with "MASCULINO" in column: 'Sexo'
    lesiones_victimas.loc[lesiones_victimas['Sexo'].str.lower() == "varon".lower(), 'Sexo'] = "MASCULINO"
    # Replace all instances of "Mujer" with "FEMENINO" in column: 'Sexo'
    lesiones_victimas.loc[lesiones_victimas['Sexo'].str.lower() == "Mujer".lower(), 'Sexo'] = "FEMENINO"
    # Replace all instances of "Mujer " with "FEMENINO" in column: 'Sexo'
    lesiones_victimas['Sexo'] = lesiones_victimas['Sexo'].str.replace("Mujer ", "FEMENINO", case=False, regex=False)
    # Replace all instances of "mujer" with "FEMENINO" in column: 'Sexo'
    lesiones_victimas.loc[lesiones_victimas['Sexo'].str.lower() == "mujer".lower(), 'Sexo'] = "FEMENINO"
    # Replace all instances of "sd" with "SD" in column: 'Sexo'
    lesiones_victimas['Sexo'] = lesiones_victimas['Sexo'].str.replace("sd", "SD", case=False, regex=False)
    return lesiones_victimas

lesiones_victimas= clean_data(lesiones_victimas.copy())
lesiones_victimas.head()
```

creamos una función para calcular si el siniestro ocurrió cerca de un semáforo. 

```python
from geopy.distance import geodesic

def calcular_distancia(siniestro, semaforo):
  """Calcula la distancia entre un siniestro y un semaforo urbano."""
  try:
    coordenadas_siniestro = (float(siniestro[0]), float(siniestro[1]))
    coordenadas_semaforo = (float(semaforo[0]), float(semaforo[1]))
  except ValueError:
    return None
  return geodesic(coordenadas_siniestro, coordenadas_semaforo).meters

def dentro_de_rango(siniestro, semaforos, rango):
  """Devuelve True si el siniestro está dentro del rango de distancia del semaforo."""
  for semaforo in semaforos:
    distancia = calcular_distancia(siniestro, semaforo)
    if distancia is None:
      return False
    if distancia <= rango:
      return True
  return False

valores_rango = [dentro_de_rango(siniestro, df_semaforos[['Latitud', 'Longitud']].values.tolist(), 30) for siniestro in siniestros_hechos[['Latitud', 'Longitud']].values.tolist()]
siniestros_hechos["dentro_rango_30m"] = [valor if valor is not None else False for valor in valores_rango]

siniestros_hechos.head()
```
